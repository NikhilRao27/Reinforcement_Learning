# ENVIRONMENT

```python
environment_name = 'CartPole-v1'
env = gym.make(environment_name,render_mode="human")
```

```python
#initially playing the game , by generating the random action.

episodes = 10
for episode in range(1,episodes+1):
    state = env.reset()
    done = False
    score =0 
    
    while not done :
        env.render()
        action = env.action_space.sample()
        n_state,reward,done,trunc,info =env.step(action)
        score = score + reward
    print('episode:{} score = {}'.format(episode,score))
    
    #env.close()
        
    

```

## TRAINING- PPO MODEL

```python
log_path = os.path.join("Downloads/Nikhil_learning/projects/reinforcement_learning/Logs")
```

```python
log_path
```

```python
# defininng the env and model
env = gym.make(environment_name,render_mode="human")
env=DummyVecEnv([lambda:env])
model = PPO('MlpPolicy',env,verbose = 1, tensorboard_log=log_path)
#stablebaselines versions define the available policys.
#verbose=0: This is often used when you don't want any output during training. It keeps the training process silent.

#verbose=1: This is a moderate level of verbosity. It may provide you with essential information like the current training episode number and some basic statistics.

#verbose=2 or higher: These higher values typically provide more detailed information about the training progress, including additional statistics, debug information, or logging related to the algorithm's inner workings.
```

```python
##PPO?? - CAN be used to define the function
```

```python
model.learn(total_timesteps=20000)
```
